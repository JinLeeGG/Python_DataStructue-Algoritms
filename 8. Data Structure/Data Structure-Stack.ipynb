```python
# 리스트를 통한 스택 구현

class StackWithList:
    def __init__(self):
        self.stack = []
        self.top_idx = None

    def is_empty(self):
        return len(self.stack) == 0

    def push(self, data):
        self.stack.append(data)

        if len(self.stack) == 1:
            self.top_idx = 0

        else:
            self.top_idx += 1

    def pop(self):
        if self.is_empty():
            raise IndexError('빈 스택입니다')

        data = self.stack[self.top_idx]
        del self.stack[self.top_idx]
        self.top_idx -= 1
        return data


    # top data 반환
    def peek(self):
        return self.stack[self.top_idx]

    def show(self):
      for i in self.stack[::-1]:
          print(i)

```

```python
stack = StackWithList()
```

```python
stack.push(10)
stack.push(20)
stack.push(30)
stack.push(40)
```

```python
stack.show()
```

```python
stack.peek()
```

```python
stack.pop()
stack.show()
```
