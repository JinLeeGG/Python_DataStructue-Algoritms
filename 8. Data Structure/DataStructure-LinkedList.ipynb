```python
# 연결리스트의 ADT
class LinkedList:
    # Node 내부 클래스
    class Node:
        def __init__(self, data = None, link = None):
            self.data = data
            self.link = link

    def __init__(self):
        self.head = None
        self.length = 0

    # show
    def show(self):
        current = self.head
        while current != None:
            print(current.data, end=' -> ')
            current = current.link
        print()

    # 데이터 삽입
    # 1) 맨 앞에 삽입
    def insert_first(self, data):
        new_node = self.Node(data, self.head)
        self.head = new_node
        self.length += 1

    # 2) 맨 뒤에 삽입
    def insert_last(self, data):
        # 리스트에 요소가 없을 때
        if self.head == None:
            self.insert_first(data)
            return

        # 리스트에 요소가 있을 때
        current = self.head
        while current.link != None:
            current = current.link

        new_node = self.Node(data)
        current.link = new_node
        self.length += 1

    # 3) 중간에 삽입
    def insert(self, idx, data):
        if idx < 0 or idx > self.length:
            print('인덱스 범위 벗어남')
            return

        if idx == 0:
            self.insert_first(data)
            return

        # 삽입할 위치 이전 노드
        current = self.head
        for i in range(idx - 1):
            current = current.link

        new_node = self.Node(data, current.link)
        current.link = new_node
        self.length += 1

    # 삭제
    # 1) 맨 앞에 데이터를 삭제
    def remove_first(self):
        if self.head == None:
            return

        self.head = self.head.link
        self.length -= 1

    # 2) 중간에 데이터를 삭제
    def remove(self, idx):
        if idx < 0 or idx > self.length:
            raise IndexError('인덱스 범위를 벗어남')

        if idx == 0:
            self.remove_first()
            return

        current = self.head
        for i in range(idx):
            prev = current
            current = current.link

        prev.link = current.link
        self.length -= 1

    def update(self, idx, data):
        if idx < 0 or idx > self.length:
            print('인덱스 오류')
            return

        current = self.head
        for i in range(idx):
            current = current.link
        current.data = data

    # 검색
    def select(self, idx):
        if idx < 0 or idx >= self.length:
            print('인덱스 오류')
            return

        current = self.head
        for i in range(idx):
            current = current.link
        return current.data

    # 총 개수
    def size(self):
        return f"현재 길이 : {self.length}"
```

```python
# 연결리스트
li1 = LinkedList()

li1.insert_first(2)
li1.insert_first(1)

li1.insert(0, 10)
li1.insert(1, 30)

li1.insert_last(100)
print(li1.size())

li1.show()
```

## 원형리스트


```python
# 단순 원형 리스트

class Node:
    def __init__(self):
        self.data = None
        self.link = None

# node1
node1 = Node()
node1.data = '짱구'
node1.link = node1    # 첫 번째 원형 리스트는 자기 자신을 링크로 가르킨다.

print(node1.data)
print(node1.link)

# node2
node2 = Node()
node2.data = '철수'
node1.link = node2
node2.link = node1

# node3
node3 = Node()
node3.data = '훈이'
node2.link = node3
node3.link = node1

# [node1] -> [node2] -> [node3] -> [node1] ...
```

```python
current = node1
print(current.data, end = ' -> ')

while current.link != node1:
    current = current.link
    print(current.data, end=' -> ')
```
