```python
import requests
from bs4 import BeautifulSoup as bs

url = 'https://sports.news.naver.com/kbaseball/record/index?category=kbo'

response = requests.get(url)
html = response.text
soup = bs(html, 'html.parser') #문자열에서 태그로 바꿔주는 과정

#순위 들고오기
# strong = soup.find_all('strong')
# print(strong)

ranking_table = soup.find('tbody', id = 'regularTeamRecordList_table')
# print(ranking_table)

rows = ranking_table.find_all('tr')
# print(rows)

# 순위, 팀 이름을 추출
for i in rows:
  th = i.find_all('th')[0].text.strip()
  # print(th.text.strip() + '위') #스트립은 공백제거
  div = i.find_all('div')[0].text.strip()
  st = i.find_all('strong')[1].text.strip()

  print(f"{th}위, {div}, 승률 : {st}")


```

```python
with open('./ranking_text.txt', 'w', encoding = 'utf-8') as f:
    for i in rows:
          th = i.find_all('th')[0].text.strip()
          div = i.find_all('div')[0].text.strip()
          st = i.find_all('strong')[1].text.strip()
          f.write(f"{th}위, {div}, 승률 : {st} \n")
    f.close()

print('랭킹 정보 저장을 완료했습니다.')
```

```python
import requests
from bs4 import BeautifulSoup as bs

url = 'https://www.melon.com/chart/index.htm'

headers = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36"} #컴퓨터 마다 다 다름
response = requests.get(url, headers = headers)

# print(response)

soup = bs(response.text, 'html.parser')

songs = soup.find_all('tr', class_='lst50')
# print(songs)

for i in songs:
  #순위 추출
  ranks_number = i.find_all('span', class_='rank')[0].text +'위'
  #print(ranks_number)

  #노래제목 추출
  title = i.find_all('a')[2].text
  print(f"{ranks_number}, {title}")

```
