```python
import requests
from bs4 import BeautifulSoup as bs

url = 'https://sports.news.naver.com/kbaseball/record/index?category=kbo'

response = requests.get(url)
html = response.text
soup = bs(html, 'html.parser') #ë¬¸ìì—´ì—ì„œ íƒœê·¸ë¡œ ë°”ê¿”ì£¼ëŠ” ê³¼ì •

#ìˆœìœ„ ë“¤ê³ ì˜¤ê¸°
# strong = soup.find_all('strong')
# print(strong)

ranking_table = soup.find('tbody', id = 'regularTeamRecordList_table')
# print(ranking_table)

rows = ranking_table.find_all('tr')
# print(rows)

# ìˆœìœ„, íŒ€ ì´ë¦„ì„ ì¶”ì¶œ
for i in rows:
  th = i.find_all('th')[0].text.strip()
  # print(th.text.strip() + 'ìœ„') #ìŠ¤íŠ¸ë¦½ì€ ê³µë°±ì œê±°
  div = i.find_all('div')[0].text.strip()
  st = i.find_all('strong')[1].text.strip()

  print(f"{th}ìœ„, {div}, ìŠ¹ë¥  : {st}")


```
