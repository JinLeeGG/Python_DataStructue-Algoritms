## column, row를 동시에 선택하기

## Boolean selection으로 row 선택가능
    -numpy와 동일한 방식으로 해당 조건에 맞는 row 선택이 가능하다.

```python
# 데이터 분석 1
# 주어진 데이터에서 나이가 30대이면서 1등 객석에 탑승한 승객을 선택

df.head()
```

```python
df.Pclass.value_counts()

# we need to find 216 datas
```

```python
pclass_mask = df['Pclass'] == 1
df[pclass_mask]
```

```python
age_mask = (df.Age >= 30) & (df.Age < 40)
df[age_mask & pclass_mask]
```

```python
df[age_mask & pclass_mask][['Pclass', 'Name', 'Age']]
```

```python
df[age_mask & pclass_mask][:10]
```

```python
# 데이터 분석2
# 3등급 객실의 생존자 수는?
# 결과 119명

df = load_tatinic()
df
```

```python
pclass_mask = df['Pclass'] == 3
survived_mask = df['Survived'] == 1

count = len(df[pclass_mask & survived_mask])
print(count)
```
