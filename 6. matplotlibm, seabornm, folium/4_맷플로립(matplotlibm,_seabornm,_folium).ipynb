## Folium
- 파이썬에서 대화형 지도(interactive map)를 생성할 수 있다.
특히 지리데이터(Geospatial Data) 시각화에 최적화 되어 있으며,
복잡한 지도를 간단한 코드로 구현할 수 있도록 도와주는 라이브러리이다.
- https://python-visualization.github.io/folium/latest/

```python
import folium
```

```python
# 지도생성
# location : 중심점
# zoom_start : 지도 확대 범위
map_folium = folium.Map(location=[37.56652479270633, 126.9779129116262], zoom_start=15)

# popup : 클릭 시 생성할 팝업
popup = folium.Popup('서울특별시청', max_width=200)
folium.Marker(location = [37.56652479270633, 126.9779129116262], popup=popup).add_to(map_folium)

# 위도와 경도를 검색해서 찍어보기
# https://www.google.co.kr/maps/?hl=ko
# 37.500029, 127.035452
popup = folium.Popup('코리아IT아카데미 강남점', max_width=200)
folium.Marker(location = [37.500029, 127.035452], popup=popup, icon=folium.Icon(icon='star', color='red')).add_to(map_folium)


map_folium
```

```python
data = shop_eat_korean_food.iloc[:100]

food_map = folium.Map(location=[data['위도'].mean(), data['경도'].mean()], zoom_start=12)

for i in data.index:
    food_name = data.loc[i, '상호명'] + " :" + data.loc[i, '도로명주소']
    # print(food_name)
    popup = folium.Popup(food_name, max_width=500)
    folium.Marker(location = [data.loc[i, '위도'], data.loc[i, '경도']], popup=popup).add_to(food_map)

food_map
```
